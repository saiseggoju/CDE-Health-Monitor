This project monitors **data freshness** and **validity** for Customer Data Elements (CDEs).
It helps teams identify stale or inconsistent data early, ensuring high-quality analytics pipelines.

Key Objectives:
- Detect missing or outdated records
- Visualize data quality metrics for CDEs
- Support ETL/monitoring automation

2.Setup and envirnoment


```python
# Import required libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os

# Ensure required folder exists
os.makedirs("processed", exist_ok=True)
```

3.Generate Sample Data

```python
# --- Create sample freshness data ---
freshness_data = pd.DataFrame({
    "CDE": ["Customer_ID", "Address", "Phone", "Email", "DOB"],
    "Freshness_Days": [2, 5, 1, 4, 3]
})

# --- Create sample validity data ---
validity_data = pd.DataFrame({
    "CDE": ["Customer_ID", "Address", "Phone", "Email", "DOB"],
    "Invalid_Percentage": [1.2, 3.5, 0.8, 2.0, 1.0]
})

# --- Save to processed folder ---
freshness_data.to_excel("processed/CDE_Health_Report.xlsx", index=False)
validity_data.to_excel("processed/discrepancy_report.xlsx", index=False)

print("✅ Excel files created inside 'processed' folder!")
```

📊 4. Load Data & Visualize Metrics

```python
# Load the processed Excel data
freshness_data = pd.read_excel("processed/CDE_Health_Report.xlsx")
validity_data = pd.read_excel("processed/discrepancy_report.xlsx")

# --- Visualization 1: Data Freshness ---
plt.figure(figsize=(8,5))
sns.barplot(x='CDE', y='Freshness_Days', data=freshness_data, palette='Blues_d')
plt.title("📅 Data Freshness by CDE")
plt.xticks(rotation=45)
plt.tight_layout()
plt.savefig("Freshness_by_CDE.png")
plt.show()

# --- Visualization 2: Missing vs Valid Records ---
plt.figure(figsize=(8,5))
sns.barplot(x='CDE', y='Invalid_Percentage', data=validity_data, palette='Reds')
plt.title("❌ Missing vs Valid Records by CDE")
plt.xticks(rotation=45)
plt.tight_layout()
plt.savefig("Missing_vs_Valid.png")
plt.show()

print("✅ Dashboard charts generated successfully!")


```
